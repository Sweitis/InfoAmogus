## Примеры отличия от классической математики:
* round (number[, ndigits]) - округляет число number до ndigits знаков после запятой (по умолчанию, до нуля знаков, то есть, до ближайшего целого)



```python
>>> round(1.5)
```

```python
>>> round(2.5)
```

```python
>>> round(2.65, 1)
```

```python
>>> round(2.75, 1)
```

* Десятичные дроби в Python хранятся в формате с плавающей точкой и представлены типом *float*

```python
>>> 1.
```

```python
>>> .1
```

```python
>>> 4.2
```

```python
>>> 4.2e3  # то же, что и 4.2 * 10 ** 3 == 4.2 * 1000
```

```python
>>> 4.2e-3  # то же, что и 4.2 * 10 ** (-3) == 4.2 * 0.001
```
