Тригонометрия и функция math.isclose. Согласно жизненным правилам и мировым истинам cos(pi/4) должен равняться sin(pi/4). Так ли это в Python?

```python
pi= math.pi
math.sin(pi/4) == math.cos(pi/4)
```

Нет. В вычислениях с плавающей точкой такое происходит. Но нам не всегда важна точность до 120-го знака после запятой. Поэтому можно воспользоваться функцией math.isclose, возвращающее True в случае если два числа близки друг к другу. По умолчанию рассматриваются только 9 знаков после запятой, но это значение всегда можно изменить по желанию.

```python
sin = math.sin(pi/4)
cos = math.cos(pi/4)
math.isclose(sin, cos)
```
